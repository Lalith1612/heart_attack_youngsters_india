```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz, plot_tree
import graphviz
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
```

```python
df.sample()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop("Heart Attack Likelihood", axis=1),  
    df["Heart Attack Likelihood"],               
    test_size=0.2,                               
    random_state=0)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)

print("Decision Tree accuracy",accuracy_score(y_test,y_pred))
```

```python
# Fit a Decision Tree model
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)  # Assuming X_train and y_train are your training features and target variable

# Convert the feature names Index to a list
feature_names = X_train.columns.tolist()

# Plot the Decision Tree
plt.figure(figsize=(20, 10))  # Adjust the size to fit the tree better
plot_tree(model, filled=True, feature_names=feature_names, class_names=[str(i) for i in model.classes_], rounded=True, fontsize=12)
plt.show()
```
